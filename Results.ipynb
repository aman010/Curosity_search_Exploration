```python
import pandas as pd
import matplotlib.pyplot as plt 

import ast
```

```python
# ------- Load CSV --------
df = pd.read_csv("episodie_scores_0.8_RND_run_1.csv")  # <-- replace with real path

# Convert list strings to python lists
df['score'] = df['score'].apply(lambda x: ast.literal_eval(x))

# Expand score into columns
scores = pd.DataFrame(df['score'].tolist())
scores.columns = [f"agent_{i}" for i in scores.columns]

df = pd.concat([df[['episode']], scores], axis=1)

# ------- Compute confidence --------
df['mean'] = df[[c for c in df.columns if c.startswith("agent")]].mean(axis=1)
df['std'] = df[[c for c in df.columns if c.startswith("agent")]].std(axis=1)
df['ci_low'] = df['mean'] - 1.96 * df['std']
df['ci_high'] = df['mean'] + 1.96 * df['std']

# ------- Plot --------
plt.figure(figsize=(10,5))

# Shaded CI band
plt.fill_between(df.index, df['ci_low'], df['ci_high'], alpha=0.25, label="Confidence Interval")

# Plot individual agent traces
for col in [c for c in df.columns if c.startswith("agent")]:
    plt.plot(df.index, df[col], linestyle='--', alpha=0.5, label=col)

# Plot mean
plt.plot(df.index, df['mean'], linewidth=2, label="Mean Score")

plt.xlabel("Time Step")
plt.ylabel("Reward")
plt.title("Out of Bound 0.8 RND")
plt.legend()
plt.grid()
plt.show()
```

```python
##### consolidated_results
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Nov 27 12:06:05 2025

@author: qb
"""


import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ast

RESULTS_DIR ="/home/qb/Starcraft-II-Deep-Reinforcement-Learning/results/"

EXPERIMENTS = {
    "Simple_approach_augmented + Curiosity 0.1": "Simple_approach/aguemted/multiagent_agumented_0.1-agent.csv",
    "Simple_approch_augmented + Curiosity 0.8": "Simple_approach/aguemted/multiagent_agumented_0.8-agent.csv",
    "Simple_approach_non_augmented  + Curiosity 0.1": "Simple_approach/Non_agumented/multi_agent_0.1_not_agumented_non_RND.csv",
    "Simple_apprach_non_augmented + Curiosity 0.8": "Simple_approach/Non_agumented/multi_agent_0.8_not_agumented_non_RND.csv",
    "Feature_base_augmented + Curiosity 0.1": "feature_based/agumented/agumented_0.1_feature_based_with_action.csv",
    "Feature_based_augmented + Curiosity 0.8": "feature_based/agumented/agumented_0.8_feature_based_with_action.csv",
    "Feature_based_non_augmented + Curiosity 0.1": "feature_based/non_agumented/episodie_scores_0.1_feature_based_with_action.csv",
    "Feature_baded_non_augmented + Curiosity 0.8": "feature_based/non_agumented/episodie_scores_0.8_RND_with_action.csv",
    "RND_augmented + Curioisty 0.1": "RND/agument_RND_multi_agent_0.1.csv",
    "RND_augmented + Curosity 0.8": "RND/agument_RND_multi_0.8-agent.csv",
    "RND_non_augmented + Curosity 0.1": "RND/episodie_scores_0.1_RND.csv",
    "RND_non_augmented + Curosity 0.8": "RND/outbound/episodie_scores_0.8_RND_run_1",
}


def load_and_parse(path):
    """
    Loads result files where each row is:
        score
        [-0.19, -0.20]
    Converts column 'score' to numerical columns.
    """
    df = pd.read_csv(path, on_bad_lines='skip')
    df_ = pd.DataFrame()
    # Convert string list  Python list  separate columns
    df["score_list"] = df["score"].apply(lambda x: ast.literal_eval(x))
    agent1 = df[df['agent'] == 0]['score_list'].apply(lambda x: x[0]).reset_index(drop=True)
    agent2 = df[df['agent'] == 1]['score_list'].apply(lambda x: x[1]).reset_index(drop=True)
    df_['agent1'] = agent1
    df_['agent2'] = agent2
    # t  t+1 shifting
#     df_["agent1_t+1"] = df_['agent1'].shift(-1)
#     df_["agent2_t+1"] = df_["agent2"].shift(-1)

    # Gap between agents
#     df_["gap"] = df_["agent1"] - df_["agent2"]
    
    return df_



def summarize(df_):
    """Returns clean metrics for research papers."""
    
    return {
        "mean_agent1": df_["agent1"].mean(),
        "mean_agent2": df_["agent2"].mean(),
        "std_agent1": df_['agent1'].std(),
        "std_agent2": df_['agent2'].std()
    }


def consolidate_all():
    summaries = {}
    
    for name, file in EXPERIMENTS.items():
        path = os.path.join(RESULTS_DIR, file)
        if os.path.exists(path):
            print(f"[Loading] {path}")

            df = load_and_parse(path)
            summaries[name] = summarize(df)

            # Save expanded version
            df.to_csv(path.replace(".csv", "_expanded.csv"), index=False)

    # Create the final dataframe with summaries
    final = pd.DataFrame(summaries).T
    
    # Save the final consolidated results as CSV
    final.to_csv("final_consolidated_results.csv")
    print("\nSaved  final_consolidated_results.csv")

    # Now, generate the Markdown version of the table for README
    markdown_table = final_to_markdown(final)
    
    # Save the markdown table to a .md file (README.md or other)
    with open("consolidated_results_table.md", "w") as f:
        f.write(markdown_table)
    
    print("\nSaved  consolidated_results_table.md")
    
    return final


def final_to_markdown(final_df):
    """Converts the final DataFrame to a markdown table."""
    # Initialize the markdown table string
    markdown = "| Experiment | Mean Agent 1 | Mean Agent 2 | Std Agent 1 | Std Agent 2 |\n"
    markdown += "|------------|--------------|--------------|-------------|-------------|\n"

    # Iterate over the rows of the DataFrame
    for experiment, metrics in final_df.iterrows():
        markdown += f"| {experiment} | {metrics['mean_agent1']:.3f} | {metrics['mean_agent2']:.3f} | {metrics['std_agent1']:.3f} | {metrics['std_agent2']:.3f} |\n"
    
    return markdown



if __name__ == "__main__":
    table = consolidate_all()
    print(final_to_markdown(table))
```
